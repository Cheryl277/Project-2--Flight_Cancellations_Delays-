### 1.2
- **% of Cancelled flights by Airline**
- **% of cancelled by Origin_aiports**


```python
#Total number of unique Airports
df['ORIGIN_AIRPORT'].nunique()
```

```python
#Total Number of flight taken from origin airport
airport_count = df['ORIGIN_AIRPORT'].value_counts().to_frame().reset_index().rename(columns={'index': 'ORIGIN_AIRPORT', 'ORIGIN_AIRPORT': 'Flight_Count'})
airport_count.head(10)
```

```python
#Total number of cancelled flights from origin airport
canc = df.groupby('ORIGIN_AIRPORT')['CANCELLED'].sum().sort_values(ascending=False).to_frame().reset_index().rename(columns={'index':'ORIGIN_AIRPORT','CANCELLED':'Cancelled'})
canc.head(10)
```

```python
cn_df = pd.merge(canc,airport_count, on='ORIGIN_AIRPORT')
cn_df.head(10)
```

```python
cn_df['Percentage_of_cancelled_flighs'] = cn_df['Cancelled']/cn_df['Flight_Count']*100
cn_df
```

```python
#Total numner of unique Airline
df['AIRLINE'].nunique()
```

**% of cancelled by Airline**

```python
#Total number of flights for a unique airlines 
airline_count = df['AIRLINE'].value_counts().sort_values(ascending=False).to_frame().reset_index().rename(columns={'index':'AIRLINE','AIRLINE':'FLIGHT_COUNT'})
airline_count
```

```python
#Total number of cancelled flights for each airline
ar_cnc = df.groupby('AIRLINE')['CANCELLED'].sum().sort_values(ascending=False).to_frame().reset_index()
ar_cnc
```

```python
ar_df = pd.merge(airline_count,ar_cnc,on='AIRLINE')
ar_df['Percentage_of_cancelled_flighs'] = ar_df['CANCELLED']/ar_df['FLIGHT_COUNT']*100
ar_df
```

**% of Security_delay by Origin_airports**

```python
df_sd =df.groupby('ORIGIN_AIRPORT')['SECURITY_DELAY'].apply(lambda x: (x > 0).sum()).sort_values(ascending=False).to_frame().reset_index().rename(columns={'index':'ORIGIN_AIRPORT','SECURITY_DELAY':'Delay'})
df_sd
```

```python
airport_count
```

```python
# Percentage of delayed flights due to security
sd =pd.merge(airport_count,df_sd,on='ORIGIN_AIRPORT')
sd['Percentage_of_cancelled_flights'] = sd['Delay']/sd['Flight_Count']*100
sd
```

```python
# Counts of airline delayed flights
df_ar =df.groupby('AIRLINE')['AIRLINE_DELAY'].apply(lambda x: (x > 0).sum()).sort_values(ascending=False).to_frame().reset_index().rename(columns={'index':'AIRLINE','AIRLINE_DELAY':'Delay'})
df_ar
```

```python
airline_count.head(5)
```

```python
arrd = pd.merge(airline_count,df_ar,on='AIRLINE')
arrd['Percentage_of_cancelled_flights'] = arrd['Delay']/arrd['FLIGHT_COUNT']*100
arrd.sort_values(by = 'Percentage_of_cancelled_flights', ascending=False)
```

```python
# Count of delayed flights by weather
wt_df = df.groupby('ORIGIN_AIRPORT')['WEATHER_DELAY'].apply(lambda x: (x > 0).sum()).sort_values(ascending=False).to_frame().reset_index().rename(columns={'index':'ORIGIN_AIRPORT','WEATHER_DELAY':'Delay'})
wt_df
```

```python
# Percentage of cancelled flights by weather delay
mwt_df = pd.merge(airport_count,wt_df,on='ORIGIN_AIRPORT')
mwt_df['Percentage_of_cancelled_flights'] = mwt_df['Delay']/mwt_df['Flight_Count']*100
mwt_df
```
